# read orders from db

```python

import os
import sys

import matplotlib.pyplot as plt
import pandas as pd
from loguru import logger

# import local module
from try_ccxt.core.service import BinanceTrader

# insert current path
sys.path.append(os.getcwd())


def analyze_trade_history(symbol="DOT/USDT", side="buy"):
    api_key = "api_key"
    api_secret = "api_secret"
    bt = BinanceTrader(api_key=api_key, api_secret=api_secret)

    # result 
    qs = bt.dao.query_orders(symbol=symbol, side=side)

    logger.info(f"{qs}")

    # list:
    df = pd.DataFrame(list(qs.dicts()))
    print(df)

    # save to csv
    df.to_csv(f"tmp/dot_usdt_orders.csv", index=False)

    # df 切片 20 条
    df = df[-100:]

    # 根据时间, 绘图, 把下单的订单都画出来
    # df.set_index('datetime', inplace=True)

    # utc-8 时间转换
    df['datetime'] = pd.to_datetime(df['datetime'], utc=True).dt.tz_convert('Asia/Shanghai')
    df['datetime'] = pd.to_datetime(df['datetime']).dt.strftime("%Y-%m-%d %H:%M:%S")
    # 绘制图表
    plt.figure(figsize=(20, 6))
    plt.plot(df['datetime'], df['price'], marker='o', color='red' if side == 'buy' else 'green')
    plt.xlabel('Time', color='blue')

    plt.xticks(rotation=60)  # 旋转45度
    plt.ylabel('Price: USD')
    plt.title(f'{symbol} - {side} Order History')
    plt.grid(True)
    plt.show()


def main():
    analyze_trade_history(symbol="DOT/USDT", side="buy")
    analyze_trade_history(symbol="DOT/USDT", side="sell")
    analyze_trade_history(symbol="DOT/BUSD", side="buy")
    analyze_trade_history(symbol="DOT/BUSD", side="sell")


if __name__ == '__main__':
    main()
```
