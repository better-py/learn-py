```python
import mplfinance as mpf
import pandas as pd

# 创建日期索引
index_date = pd.DatetimeIndex(['20210906', '20210907', '20210908', '20210909'])
# 随便创建数据
data = pd.DataFrame({
    'Open': [1, 2, 3, 4], 
    'High': [2, 3, 4, 5], 
    'Low' : [0.5, 1, 2, 2],
    'Close': [2, 3, 4, 3]
}, index=index_date)
# 调用plot函数 进行绘图
mpf.plot(data)
```

```python
mpf.plot(data,type='candle')
```

```python
mpf.plot(data,type='line')
```

```python
mpf.plot(data, type='candle', mav=2)
```

```python
mpf.plot(data, type='candle', mav=(2, 3))
```

```python
import akshare as ak
import mplfinance as mpf
import pandas as pd

# 获取上证指数每日的变化数据 OHLC
sz_index = ak.stock_zh_index_daily(symbol="sh000001")
data1 = sz_index.loc['2020-01-01 00:00:00+00:00':'2023-09-13 00:00:00+00:00']
print(data1)
data1.index = pd.to_datetime(data1.index)
print(data1)

mpf.plot(data1, type='candle', mav=(200, 300, 350), volume=True)
```

```python
data2 = sz_index.loc['2021-01-01 00:00:00+00:00':'2024-09-13 00:00:00+00:00']
print(data2)
data2.index = pd.to_datetime(data2.index)

if not data2.empty:
    mpf.plot(data2, type='candle', mav=(30, 60, 120), volume=True)
else:
    print("No data available for the specified date range")
```

```python
data3 = sz_index.loc['2021-04-01 00:00:00+00:00':'2024-09-13 00:00:00+00:00']

print(data3)
data3.index = pd.to_datetime(data3.index)


if not data3.empty:
    mpf.plot(data3, type='candle', mav=(5, 10, 20), volume=True)
else:
    print("No data available for the specified date range")
```
